```python
!sudo apt-get update > /dev/null
!sudo apt-get install -y tesseract-ocr poppler-utils > /dev/null
!pip install -q pytesseract pdf2image gTTS PyPDF2

import pytesseract
from pdf2image import convert_from_path
from gtts import gTTS
from PyPDF2 import PdfReader
from google.colab import files
import os

def extract_text_smart(pdf_path):
    print(f"\nProcessing {pdf_path}...")
    text_content = ""

    try:
        reader = PdfReader(pdf_path)
        total_pages = len(reader.pages)

        for i, page in enumerate(reader.pages):
            page_text = page.extract_text()

            if not page_text or len(page_text) < 50:
                print(f"   Page {i+1} is image-based. Using OCR...")
                images = convert_from_path(pdf_path, first_page=i+1, last_page=i+1)
                for img in images:
                    ocr_text = pytesseract.image_to_string(img)
                    text_content += ocr_text + " "
            else:
                print(f"   Page {i+1} read successfully.")
                text_content += page_text + " "

    except Exception as e:
        print(f"Error: {e}")
        return None

    return text_content

print("\nUpload your PDF file:")
uploaded = files.upload()

if uploaded:
    filename = next(iter(uploaded))
    full_text = extract_text_smart(filename)

    if full_text and len(full_text.strip()) > 0:
        print("\nGenerating Audio...")
        audio_filename = filename.replace(".pdf", ".mp3")

        tts = gTTS(text=full_text, lang='en', slow=False)
        tts.save(audio_filename)

        print(f"Done! Saved as: {audio_filename}")
        files.download(audio_filename)
    else:
        print("Could not extract text.")
else:
    print("No file uploaded.")
```
